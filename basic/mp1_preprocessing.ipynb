```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 데이터셋 불러오기
file_path = 'C:\\Users\\co279\\mp.csv'
data = pd.read_csv(file_path)

columns = ['last_name, first_name', 'player_id']

# 필요 없는 컬럼 제거
data = data.drop(columns=columns)

# 결측치 처리 (예: 평균값으로 대체)
data = data.fillna(0)

# 종속변수와 독립변수 분리
target = 'p_era'
features = data.columns.drop([target, 'year'])

# 독립변수만 선택
X = data[features]

# VIF 계산 함수
def calculate_vif(df):
    vif = pd.DataFrame()
    vif["feature"] = df.columns
    vif["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return vif

# VIF 계산
vif_data = calculate_vif(X)

# Pearson 상관계수 계산
corr_matrix = X.corr().abs()

# 상관계수가 0.8 미만인 변수들만 선택
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] >= 0.8)]
X_reduced = X.drop(columns=to_drop)

# VIF가 10 미만인 변수들만 선택
vif_data_reduced = calculate_vif(X_reduced)
vif_data_final = vif_data_reduced[vif_data_reduced["VIF"] < 10]

# 선택된 변수들
selected_features = vif_data_final["feature"].tolist()

print("선택된 변수들:", selected_features)
print(vif_data_final)
```
